```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eye_classifier = cv2.CascadeClassifier('Models/haarcascade_eye.xml')
```

```python
img = cv2.imread('Images/Sheen.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
def eyes_detect(fix_img):
    eyes_rect = eye_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in eyes_rect:
        cv2.rectangle(fix_img, (x,y),(x+w, y+h),(234,0,0),3)
        
    return fix_img
```

```python
result = eyes_detect(fix_img)
plt.imshow(result)
```
