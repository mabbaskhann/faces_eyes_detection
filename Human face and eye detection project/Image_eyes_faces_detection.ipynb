```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_classifier = cv2.CascadeClassifier('Models/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Models/haarcascade_eye.xml')
```

```python
img = cv2.imread('Images/Sheen.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
faces = face_classifier.detectMultiScale(img, 1.3,5)
```

```python
def eyes_faces_detect(fix_img):
    faces_rect = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in faces_rect:
        cv2.rectangle(fix_img, (x,y),(x+w, y+h),(234,0,0),3 )
        eyes_rect = eye_classifier.detectMultiScale(fix_img)
        for (ix,iy,iw,ih) in eyes_rect:
            cv2.rectangle(fix_img, (ix,iy), (ix+iw, iy+ih),(230,0,0),3)
            
    return fix_img
```

```python
result = eyes_faces_detect(fix_img)
plt.imshow(result)
```
